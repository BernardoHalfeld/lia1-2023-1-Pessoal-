## 4) INTRODUÇÃO AOS DADOS (TEORIA)
A cada dia, uma enorme quantidade de dados é gerada de diferentes fontes. Um conjunto de dados é formado por objetos que representam algo físico (cadeira) ou abstrato (diagnóstico), sendo descrito por seus atributos. Deve-se realizar a caracterização dos dados e a exploração inicial com estatística e visualização básicas, conhecido como pré-processamento.

O pré-processamento pode demandar tarefas de integração, eliminação manual, amostragem, balanceamento, limpeza, redução da dimensionalidade, transformação de dados etc., estando essas tarefas associadas à Mineração de Dados (MD), a qual foca no processo em si, em como realizar essas tarefas.

A MD busca extrair conhecimento novo, útil e relevante, gerando as bases para a Ciência de Dados (CD), a qual foca no impacto e na relevância do conhecimento.

**Em resumo, MD (processo) e CD (dados)!**

Antes de aplicar algoritmos de AM a um conjunto de dados, é importante que os dados sejam analisados. Essa análise, que pode ser realizada por técnicas estatísticas e de visualização, permite uma melhor compreensão da distribuição dos dados e pode dar suporte à escolha de formas de abordar o problema.

## 5) EXERCÍCIO PROPOSTO

* #### Gráfico de barras VERTICAIS utilizando matplotlib para estimar o grupo de pessoas "doente" e "saudável"

```python
df_dataset = pd.read_csv('hospital.txt', sep = ',', index_col=None)

contSaudaveis = len(df_dataset[df_dataset['Diagnóstico']=='Saudável'].value_counts(dropna=False))
contDoentes = len(df_dataset[df_dataset['Diagnóstico']=='Doente'].value_counts())
valores = (contSaudaveis, contDoentes)
atributos = ('Saudáveis', 'Doentes')

plt.bar(atributos, valores, color = ['green', 'red'], width = 0.4)
plt.title('Diagnósticos')
plt.show()
```

* #### Gráfico de barras HORIZONTAIS utilizando matplotlib para estimar o grupo de pessoas "doente" e "saudável" por time

```python
df_dataset = pd.read_csv('hospital.txt', sep = ',', index_col = None)

tabelaDoentes = df_dataset[(df_dataset['Diagnóstico']=='Doente') & (df_dataset['Time'])]
tabelaDoentes.groupby("Time").count().plot.barh(y = 'Diagnóstico', width = 0.5, color='red', title = 'Time de coração das pessoas Doentes', legend = None)

tabelaSaudaveis = df_dataset[(df_dataset['Diagnóstico']=='Saudável') & (df_dataset['Time'])]
tabelaSaudaveis.groupby("Time").count().plot.barh(y = 'Diagnóstico', width = 0.5, color='green', title = 'Time de coração das pessoas Saudáveis', legend = None)
```

* #### Gráfico de barras VERTICAL utilizando seaborn para estimar a média de temperatura entre homens e mulheres doentes e depois saudáveis


```python
df_dataset = pd.read_csv('hospital.txt', sep = ',', index_col=None)

tabelaSaudaveis = df_dataset[(df_dataset['Diagnóstico']=='Saudável') & (df_dataset['Time'])]
tabelaSaudaveis = sns.barplot(x = 'Sexo',y = 'Peso', data = tabelaSaudaveis, palette = 'Blues')
plt.title('Média de altura de homens e mulheres saudáveis')
plt.show()

tabelaDoentes = df_dataset[(df_dataset['Diagnóstico']=='Doente') & (df_dataset['Time'])]
tabelaDoentes = sns.barplot(x = 'Sexo',y = 'Peso', data = tabelaDoentes, palette = 'magma')
plt.title('Média de altura de homens e mulheres doentes')
plt.show()
```

* #### Gráfico de barras HORIZONTAIS utilizando seaborn para estimar os estados de pessoas doentes

```python
df_dataset = pd.read_csv('hospital.txt', sep = ',', index_col=None)

tabelaDoentes = df_dataset[(df_dataset['Diagnóstico']=='Doente') & (df_dataset['Time'])]
tabelaDoentes = sns.countplot(y = 'Estado', orient='h', data = tabelaDoentes, palette = 'mako')
plt.title('Quantas pessoas doentes por estado')
plt.show()
```
